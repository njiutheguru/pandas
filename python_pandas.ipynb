```python
# drop all the duplicates
df.drop_duplicates()
```

```python
# remove unuseful / unnecessary columns
df.drop("User_ID", axis=1, inplace=True)
```

```python
df.drop("Product_ID",axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.Stay_In_Current_City_Years.unique()
```
